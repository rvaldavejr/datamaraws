```python
import geopandas as gpd
import pandas as pd

# 1. Load the full DHS dataset
input_path = '/Users/ruben/Desktop/Thesis/TrainingData/PH_DHS_GPS/PHGE81FL/PHGE81FL.shp'
output_path = '/Users/ruben/Desktop/Thesis/TrainingData/PH_DHS_GPS/ph_dhs_sample_50.shp'

gdf = gpd.read_file(input_path)
print(f"Total Clusters Loaded: {len(gdf)}")

# 2. Randomly sample points

# Separate them
urban = gdf[gdf['URBAN_RURA'] == 'U']
rural = gdf[gdf['URBAN_RURA'] == 'R']

# Take 25 of each
sample_u = urban.sample(n=25)
sample_r = rural.sample(n=25)

# Combine them back together
balanced_sample = pd.concat([sample_u, sample_r])

# Save
balanced_sample.to_file(output_path)



print(f"Saved {len(balanced_sample)} sample points to: {output_path}")
```
